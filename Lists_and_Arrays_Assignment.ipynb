```python
# List
games = ["2k", "minecraft", "fortnite", "madden"]
print(games[3])
# Modifying
games[2] = "uncharted"
print(games)
#Adding a new game
games.append("Trove")
print(games)
# Inserting a name game at the beginning
games.insert(0, "Roblox")
print(games)
# Removing a "minecraft" out of the list
games.remove('minecraft')
print(games)
# Printing the length of the game list
print(len(games))
# A for loop that shows each color of the list in a new line
games = ['Roblox', '2k', 'uncharted', 'madden', 'Trove']
for item in games:
  print(item)
# A 3x3 grid representing every game with each slot.
games = [["MT" for i in range(3)] for i in range(3)]
# Prints the title before the rows are displayed.
print("Games Played:")
for row in games:
# Prints the games played rows.
  print(row)

# Replace the middle slot
games[1][1] = "Uncharted"
# Replace the top-left slot
games[0][0] = "Roblox"
# Replace the bottom-right slot
games[2][2] = "Trove"
# Print the updated games grid
print("\nUpdated Games Played:")
for row in games:
    print(row)
# Another title
print("\nAccessing Specific Slots:")
# Print the game in the first row and the second column.
print("First row, second column:", games[0][1])
# Print the last game in the last row and last column.
print("Last row, last column:", games[2][2])
# Step 1: Create a 2D Games Played List
games = [["Empty" for _ in range(3)] for _ in range(3)]
print("Initial Games Played:")
for row in games:
    print(row)

# Step 2: Place Games Played in the list
games[1][1] = "Uncharted"
games[0][0] = "Roblox"
games[2][2] = "Trove"

print("\nUpdated Games Played:")
for row in games:
    print(row)

# Step 3: Access Specific Slots
print("\nAccessing Specific Slots:")
print("First row, second column:", games[0][1])
print("Last row, last column:", games[2][2])

```
