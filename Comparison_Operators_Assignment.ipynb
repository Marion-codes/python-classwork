```python
# Basketball Attempts
three_point_attempts = int(input("Enter the amount of three point attempts: "))
free_throws_attempts = int(input("Enter the amount of free throw attempts: "))

print("Are the amount of shots taken equal?")
print(three_point_attempts == free_throws_attempts)

print("Are the amount of shots taken different?")
print(three_point_attempts != free_throws_attempts)

print("Did you shoot more three points?")
print(three_point_attempts > free_throws_attempts)

print("Did you shoot more free throws?")
print(three_point_attempts < free_throws_attempts)

print("Is the amount of the three point attempts greater than or equal to the free throw attempts?")
print(three_point_attempts >= free_throws_attempts)

print("Is the amount of the three point attempts less than or equal to the amount of free throw attempts?")
print(three_point_attempts <= free_throws_attempts)
```
