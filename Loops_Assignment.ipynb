```python
"""
Name: Marion Brown
Date: 6/22/26
Project: Football Game
Description: A football game with loops
"""

clock = 5

for clock_left in range(clock, 0, -1):
  if clock_left == 1:
    print(f"With only {clock_left} second left, they snap the ball!")
  else:
    print(f"{clock_left} seconds left on the play clock")

print("The ball is caught in the endzone! TOUCHDOWN!")
```
